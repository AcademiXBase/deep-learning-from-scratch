```python
import numpy as np
import matplotlib.pyplot as plt

# 2点
P = np.array([2.0, 2.0])
Q = np.array([7.0, 6.0])

fig, ax = plt.subplots(figsize=(10, 5.5))

# ---- 点 ----
ax.scatter([P[0], Q[0]], [P[1], Q[1]],
           s=220, color="#2D2A83", zorder=5)

# ---- マンハッタン経路（横 → 縦）----
# 横
ax.annotate("",
            xy=(Q[0], P[1]),
            xytext=P,
            arrowprops=dict(arrowstyle="->",
                            lw=3,
                            color="#2D2A83"))

# 縦
ax.annotate("",
            xy=Q,
            xytext=(Q[0], P[1]),
            arrowprops=dict(arrowstyle="->",
                            lw=3,
                            color="#2D2A83"))

# ---- a ----
ax.text((P[0]+Q[0])/2, P[1]-0.6,
        "a", fontsize=18, ha="center")

# ---- b ----
ax.text(Q[0]+0.4, (P[1]+Q[1])/2,
        "b", fontsize=18, va="center")

# ---- L1 ----
ax.text((P[0]+Q[0])/2 - 0.5,
        (P[1]+Q[1])/2,
        r"$L_1$",
        fontsize=22)

# ---- 見た目調整 ----
ax.set_aspect("equal")
ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 2点
P = np.array([2.0, 2.0])
Q = np.array([7.0, 6.0])

fig, ax = plt.subplots(figsize=(10, 5.5))

# ---- 点 ----
ax.scatter([P[0], Q[0]], [P[1], Q[1]],
           s=220, color="#2D2A83", zorder=5)

# ---- a（右向き）----
ax.annotate("",
            xy=(Q[0], P[1]),
            xytext=P,
            arrowprops=dict(arrowstyle="->",
                            linestyle="--",
                            lw=2.5,
                            color="#2D2A83"))

ax.text((P[0]+Q[0])/2, P[1]-0.5,
        "a", fontsize=18, ha="center")

# ---- b（上向き）----
ax.annotate("",
            xy=Q,
            xytext=(Q[0], P[1]),
            arrowprops=dict(arrowstyle="->",
                            linestyle="--",
                            lw=2.5,
                            color="#2D2A83"))

ax.text(Q[0]+0.4, (P[1]+Q[1])/2,
        "b", fontsize=18, va="center")

# ---- L2（P → Q）----
ax.annotate("",
            xy=Q,
            xytext=P,
            arrowprops=dict(arrowstyle="->",
                            lw=2.8,
                            color="#2D2A83"))

# ---- L2 ラベル（少し浮かせる）----
mid = (P + Q) / 2
direction = Q - P
normal = np.array([-direction[1], direction[0]])
normal = normal / np.linalg.norm(normal)

offset = 0.4
label_pos = mid + normal * offset

ax.text(label_pos[0], label_pos[1],
        r"$L_2$", fontsize=18)

# ---- 見た目調整 ----
ax.set_aspect("equal")
ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 2点
P = np.array([2.0, 2.0])
Q = np.array([7.0, 6.0])

a = Q[0] - P[0]
b = Q[1] - P[1]

fig, ax = plt.subplots(figsize=(10, 5.5))

color_main = "#2D2A83"

# ---- 点 ----
ax.scatter([P[0], Q[0]], [P[1], Q[1]],
           s=220, color=color_main, zorder=5)

# ---- a（点線）----
ax.annotate("",
            xy=(Q[0], P[1]),
            xytext=P,
            arrowprops=dict(arrowstyle="->",
                            linestyle="--",
                            lw=2.5,
                            color=color_main))

ax.text((P[0]+Q[0])/2, P[1]-0.5,
        "a", fontsize=18, ha="center")

# ---- b（点線）----
ax.annotate("",
            xy=Q,
            xytext=(Q[0], P[1]),
            arrowprops=dict(arrowstyle="->",
                            linestyle="--",
                            lw=2.5,
                            color=color_main))

ax.text(Q[0]+0.4, (P[1]+Q[1])/2,
        "b", fontsize=18, va="center")

# ---- L∞（最大成分を強調・青）----
if abs(a) >= abs(b):
    ax.annotate("",
                xy=(Q[0], P[1]+0.8),
                xytext=(P[0], P[1]+0.8),
                arrowprops=dict(arrowstyle="<->",
                                lw=3.5,
                                color=color_main))
    ax.text((P[0]+Q[0])/2, P[1]+1.0,
            r"$L_\infty$",
            fontsize=20, ha="center",
            color=color_main)
else:
    ax.annotate("",
                xy=(Q[0]+0.8, Q[1]),
                xytext=(Q[0]+0.8, P[1]),
                arrowprops=dict(arrowstyle="<->",
                                lw=3.5,
                                color=color_main))
    ax.text(Q[0]+1.0, (P[1]+Q[1])/2,
            r"$L_\infty$",
            fontsize=20, va="center",
            color=color_main)

# ---- 見た目調整 ----
ax.set_aspect("equal")
ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

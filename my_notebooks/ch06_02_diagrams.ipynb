```python
import numpy as np
import matplotlib.pyplot as plt

# 2点
P = np.array([2.0, 2.0])
Q = np.array([7.0, 6.0])

fig, ax = plt.subplots(figsize=(10, 5.5))

# ---- 点 ----
ax.scatter([P[0], Q[0]], [P[1], Q[1]],
           s=220, color="#2D2A83", zorder=5)

# ---- a（右向き）----
ax.annotate("",
            xy=(Q[0], P[1]),
            xytext=P,
            arrowprops=dict(arrowstyle="->",
                            linestyle="--",
                            lw=2.5,
                            color="#2D2A83"))

ax.text((P[0]+Q[0])/2, P[1]-0.5,
        "a", fontsize=18, ha="center")

# ---- b（上向き）----
ax.annotate("",
            xy=Q,
            xytext=(Q[0], P[1]),
            arrowprops=dict(arrowstyle="->",
                            linestyle="--",
                            lw=2.5,
                            color="#2D2A83"))

ax.text(Q[0]+0.4, (P[1]+Q[1])/2,
        "b", fontsize=18, va="center")

# ---- L2（P → Q）----
ax.annotate("",
            xy=Q,
            xytext=P,
            arrowprops=dict(arrowstyle="->",
                            lw=2.8,
                            color="#2D2A83"))

# ---- L2 ラベル（少し浮かせる）----
mid = (P + Q) / 2
direction = Q - P
normal = np.array([-direction[1], direction[0]])
normal = normal / np.linalg.norm(normal)

offset = 0.4
label_pos = mid + normal * offset

ax.text(label_pos[0], label_pos[1],
        r"$L_2$", fontsize=18)

# ---- 見た目調整 ----
ax.set_aspect("equal")
ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 2点
P = np.array([2.0, 2.0])
Q = np.array([7.0, 6.0])

a = Q[0] - P[0]
b = Q[1] - P[1]

fig, ax = plt.subplots(figsize=(10, 5.5))

color_main = "#2D2A83"

# ---- 点 ----
ax.scatter([P[0], Q[0]], [P[1], Q[1]],
           s=220, color=color_main, zorder=5)

# ---- a（点線）----
ax.annotate("",
            xy=(Q[0], P[1]),
            xytext=P,
            arrowprops=dict(arrowstyle="->",
                            linestyle="--",
                            lw=2.5,
                            color=color_main))

ax.text((P[0]+Q[0])/2, P[1]-0.5,
        "a", fontsize=18, ha="center")

# ---- b（点線）----
ax.annotate("",
            xy=Q,
            xytext=(Q[0], P[1]),
            arrowprops=dict(arrowstyle="->",
                            linestyle="--",
                            lw=2.5,
                            color=color_main))

ax.text(Q[0]+0.4, (P[1]+Q[1])/2,
        "b", fontsize=18, va="center")

# ---- L∞（最大成分を強調・青）----
if abs(a) >= abs(b):
    ax.annotate("",
                xy=(Q[0], P[1]+0.8),
                xytext=(P[0], P[1]+0.8),
                arrowprops=dict(arrowstyle="<->",
                                lw=3.5,
                                color=color_main))
    ax.text((P[0]+Q[0])/2, P[1]+1.0,
            r"$L_\infty$",
            fontsize=20, ha="center",
            color=color_main)
else:
    ax.annotate("",
                xy=(Q[0]+0.8, Q[1]),
                xytext=(Q[0]+0.8, P[1]),
                arrowprops=dict(arrowstyle="<->",
                                lw=3.5,
                                color=color_main))
    ax.text(Q[0]+1.0, (P[1]+Q[1])/2,
            r"$L_\infty$",
            fontsize=20, va="center",
            color=color_main)

# ---- 見た目調整 ----
ax.set_aspect("equal")
ax.set_xlim(0, 10)
ax.set_ylim(0, 8)
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import FancyArrowPatch
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 描画用の関数
def draw_neuron(ax, x, y, radius=0.3, is_active=True):
    """ニューロンを描画"""
    if is_active:
        circle = patches.Circle((x, y), radius, fill=False, edgecolor='black', linewidth=2)
        ax.add_patch(circle)
    else:
        # 消去されたニューロン（×印）
        circle = patches.Circle((x, y), radius, fill=False, edgecolor='black', linewidth=2)
        ax.add_patch(circle)
        # ×印を描画
        ax.plot([x-radius*0.6, x+radius*0.6], [y-radius*0.6, y+radius*0.6], 'k-', linewidth=2)
        ax.plot([x-radius*0.6, x+radius*0.6], [y+radius*0.6, y-radius*0.6], 'k-', linewidth=2)

def draw_connection(ax, x1, y1, x2, y2, radius=0.3):
    """ニューロン間の接続を矢印付きで描画（ニューロンの境界から始まる）"""
    # 方向ベクトルを計算
    dx = x2 - x1
    dy = y2 - y1
    distance = np.sqrt(dx**2 + dy**2)
    
    # 正規化された方向ベクトル
    if distance > 0:
        dx_norm = dx / distance
        dy_norm = dy / distance
    else:
        dx_norm, dy_norm = 0, 0
    
    # 始点と終点をニューロンの境界に調整
    start_x = x1 + radius * dx_norm
    start_y = y1 + radius * dy_norm
    end_x = x2 - radius * dx_norm
    end_y = y2 - radius * dy_norm
    
    arrow = FancyArrowPatch((start_x, start_y), (end_x, end_y),
                           arrowstyle='->', mutation_scale=15,
                           linewidth=0.8, color='black')
    ax.add_patch(arrow)

# 層の設定
layers = [1, 5, 5, 5, 1]  # 各層のニューロン数
layer_names = ['Input', 'Hidden1', 'Hidden2', 'Hidden3', 'Output']
x_positions = [1, 3, 5, 7, 9]  # 各層のx座標

# (a) 通常のニューラルネットワーク
ax1.set_xlim(0, 10)
ax1.set_ylim(-1, 6)
ax1.axis('off')
ax1.set_title('(a) Standard Neural Net', fontsize=14, fontweight='bold')

# ニューロンの描画と接続
neurons_pos = []
for layer_idx, (n_neurons, x_pos) in enumerate(zip(layers, x_positions)):
    layer_neurons = []
    y_offset = (5 - n_neurons) / 2
    for i in range(n_neurons):
        y = i + y_offset
        draw_neuron(ax1, x_pos, y)
        layer_neurons.append((x_pos, y))
    neurons_pos.append(layer_neurons)
    
    # 前の層との接続を描画
    if layer_idx > 0:
        for x1, y1 in neurons_pos[layer_idx - 1]:
            for x2, y2 in layer_neurons:
                draw_connection(ax1, x1, y1, x2, y2)

# (b) Dropout適用後のニューラルネットワーク
ax2.set_xlim(0, 10)
ax2.set_ylim(-1, 6)
ax2.axis('off')
ax2.set_title('(b) After applying dropout', fontsize=14, fontweight='bold')

# Dropoutパターン（ランダムに一部を消去）
np.random.seed(42)  # 再現性のため
dropout_pattern = [
    [True],  # Input layer
    [True, False, True, False, True],  # Hidden1
    [False, True, False, False, True],  # Hidden2
    [True, False, True, False, True],  # Hidden3
    [True]   # Output layer
]

# ニューロンの描画
neurons_pos_dropout = []
for layer_idx, (n_neurons, x_pos, active_pattern) in enumerate(zip(layers, x_positions, dropout_pattern)):
    layer_neurons = []
    y_offset = (5 - n_neurons) / 2
    for i in range(n_neurons):
        y = i + y_offset
        is_active = active_pattern[i]
        draw_neuron(ax2, x_pos, y, is_active=is_active)
        layer_neurons.append((x_pos, y, is_active))
    neurons_pos_dropout.append(layer_neurons)
    
    # 前の層との接続を描画（両方のニューロンがアクティブな場合のみ）
    if layer_idx > 0:
        for x1, y1, active1 in neurons_pos_dropout[layer_idx - 1]:
            for x2, y2, active2 in layer_neurons:
                if active1 and active2:
                    draw_connection(ax2, x1, y1, x2, y2)

plt.tight_layout()
plt.show()
```
